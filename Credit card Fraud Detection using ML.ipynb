```python
class_0 = 284315
class_1 = 492
```

```python
credit_card = pd.concat([
    data[data['class'] == 0].sample(class_1 ,replace=True),
    data[data['class'] == 1]])
```

```python
credit_card.shape
```

```python
credit_card['class'].value_counts()
```

```python
credit_card.describe()
```

Normalizing the dataset

```python
from sklearn.preprocessing import StandardScaler
scaled_data = StandardScaler()
scaled_data1 = scaled_data.fit_transform(credit_card)
```

```python
scaled_data_df = pd.DataFrame(scaled_data1, columns=credit_card.columns)
scaled_data_df
```

Train Test Split 

```python
X = scaled_data_df.drop(['class'],axis=1)
Y = scaled_data_df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
print(X_test.shape,Y_test.shape,X_train.shape,Y_train.shape)
```

Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Accuracy on training data

```python
Y_train_predict = model.predict(X_train)
accuracy = accuracy_score(Y_train,Y_train_predict)
print(accuracy)
```

Accuracy on testing data

```python
Y_test_predict = model.predict(X_test)
accuracy = accuracy_score(Y_test,Y_test_predict)
print(accuracy)
```

Classification report 


```python
print(classification_report(Y_test,Y_test_predict))
```
